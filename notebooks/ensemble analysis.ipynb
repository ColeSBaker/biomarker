```python
from biomarker.experiments import ensemble
import numpy as np

%matplotlib notebook
```

```python
MAX_TESTS = 60
MIN_TESTS = 5
REGRESSOR = 'tree'
TOTAL_EXAMPLES = 191
META_MODEL = 'tree'
```

```python
mses = []
r2s = []
coefs = []
rng = range(MIN_TESTS,MAX_TESTS)
for i in rng:
    print 'step: ', i
    mse,r_squared,coef,_ = ensemble(i, regressor=REGRESSOR, meta_model=META_MODEL, max_depth=2)
    mses.append(mse)
    r2s.append(r_squared)
    coefs.append(coef)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(rng, mses, 'bo-', linewidth=1, markersize=6)
ax.set_title('Mean squared error versus test examples')
ax.set_xlabel('Number of test examples')
ax.set_ylabel('Mean squared error')
```

```python
fig, ax = plt.subplots()

ax.plot(rng, r2s, 'bo-', linewidth=1, markersize=6)
ax.set_title('R-squared versus test examples')
ax.set_xlabel('Number of test examples')
ax.set_ylabel('R^2')
```

```python
fig, ax = plt.subplots()

train_rng = np.full(shape=len(rng), fill_value=TOTAL_EXAMPLES)
ax.plot(train_rng - rng, mses, 'ro-', linewidth=1, markersize=6)
ax.set_title('Mean squared error versus train size')
ax.set_xlabel('Number of training examples')
ax.set_ylabel('Mean squared error (test)')
```

```python
fig, ax = plt.subplots()

train_rng = np.full(shape=len(rng), fill_value=TOTAL_EXAMPLES)
ax.plot(train_rng - rng, r2s, 'ro-', linewidth=1, markersize=6)
ax.set_title('R-squared versus train size')
ax.set_xlabel('Number of training examples')
ax.set_ylabel('R^2 (test)')
```

```python
coefs = np.array(coefs)
%matplotlib inline

fig, ax = plt.subplots(coefs.shape[1], 1, sharex='col',figsize=(20, 20))

for j in range(coefs.shape[1]):
    ax[j].plot(train_rng - rng, coefs[:,j], linewidth=1, markersize=6)

# ax.set_title('Evolution of linear regression coefficients')
# ax.set_xlabel('Number of training examples')
# ax.set_ylabel('Coefficient value')

# plt.legend(range(coefs.shape[1]), loc='upper left')
```
